```python
import numpy as np
import math
import random
from Heap import MyHeap


hashmap=dict()
def Add(C,q):
    global hashmap
    t=len(C)
    b=len(C[0])
    if q not in hashmap.keys():
        temp=dict()
        for i in range(t):
            np.random.seed(i+q)
            hi=np.random.choice(range(0,b),1)[0]
            temp[i]=hi
            C[i][hi]+=1
        hashmap[q]=temp
    else:
        for i in range(t):
            hi=hashmap[q][i]
            C[i][hi]+=1
    return C
    
def Estimate(C,q):
    global hashmap
    t=len(C)
    temp=list()
    for i in range(0,t):
        hi=hashmap[q][i]
        temp.append(C[i][hi])
    return min(temp)

def TopK(stream,C,k=10):
    global hashmap
    result=dict()
    for q in stream:
        C=Add(C,q)
        f=Estimate(C,q)
        result[q]=f
    temp=sorted(result,key=result.get,reverse=True)[:k] #按值排序
    for i in range(k):
        temp[i]=(result[temp[i]],temp[i])
    return temp

def BruteForce_Count(stream,k=10):
    counter=dict()
    for q in stream:
        if q in counter.keys():
            counter[q]+=1
        else:
            counter[q]=1
    result=sorted(counter,key=counter.get,reverse=True)[:k] #按值排序
    return result,counter

def Variance(result,counter):
    s=0
    k=len(result)
    for i in result:
        s+=(i[0]-counter[i[1]])**2
    return (s/k)**0.5

def Recall(result,real):
    k=len(real)
    temp=list()
    for i in result:
        temp.append(i[1])
    recall=len(list(set(temp)&set(real)))/k
    return recall

def Lab(δ=0.001,ε=0.0015):
    data=np.load('datastream2.npy')
    n=len(data)
    t=int(math.floor(1*(math.log(n/δ))))
    b=int(math.floor(1/(ε)))
    print(t)
    print(b)
    C=np.zeros((t,b),dtype='int32')
    result=TopK(data,C)
    real,counter=BruteForce_Count(data)
    v=Variance(result,counter)
    recall=Recall(result,real)
    print(recall)
    print(v)

Lab()
```
